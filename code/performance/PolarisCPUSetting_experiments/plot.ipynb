```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
cpus_per_worker = [8, 4, 2, 1]

# Runtimes for different configurations
runtime_200_pdfs = [87, 28, 28, 28]
runtime_1024_pdfs = [111, 54, 54, 54]
runtime_24000_zips = [184, 161, 120, 95]

# Constants
n_200 = 200
n_1024 = 4
n_24000 = 24

N_200 = 200
N_1024 = 1024
N_24000 = 24000

# Calculate Avg. runtime per (runtime / (N * cpus_per_worker))
avg_runtime_200_pdfs = [rt / (n_200 ) for rt, cpw in zip(runtime_200_pdfs, cpus_per_worker)]
avg_runtime_1024_pdfs = [rt / (n_1024 ) for rt, cpw in zip(runtime_1024_pdfs, cpus_per_worker)]
avg_runtime_24000_zips = [rt / (n_24000 ) for rt, cpw in zip(runtime_24000_zips, cpus_per_worker)]

# Create a grouped bar chart
x = np.arange(len(cpus_per_worker))
width = 0.2

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width, avg_runtime_200_pdfs, width, label='200 PDFs')
ax.bar(x, avg_runtime_1024_pdfs, width, label='1024 PDFs (256 PDFs/ZIP)')
ax.bar(x + width, avg_runtime_24000_zips, width, label='24000 ZIPs (1000 PDFs/ZIP)')

# Labels and title
ax.set_xlabel('CPUs per Worker')
ax.set_ylabel('Avg. Runtime per Unit')
ax.set_title('Avg. Runtime per Unit vs. CPUs per Worker')
ax.set_xticks(x)
ax.set_xticklabels(cpus_per_worker)
ax.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
cpus_per_worker = [8, 4, 2, 1]
avg_runtime = [0.05, 0.1, 0.15, 50]  # Example data

# Create subplots
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(8, 6))

# Plot the same data on both axes but with different y-limits
ax1.bar(cpus_per_worker, avg_runtime)
ax2.bar(cpus_per_worker, avg_runtime)

# Set y-axis limits for the lower and upper subplots
ax1.set_ylim(45, 55)  # Upper part
ax2.set_ylim(0, 0.2)  # Lower part

# Remove spines and ticks between the plots
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.tick_params(labeltop=False)  # Don't put tick labels at the top of the upper plot
ax2.xaxis.tick_bottom()

# Add diagonal lines to indicate a break in the y-axis
d = .015  # How big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # Top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # Top-right diagonal

kwargs.update(transform=ax2.transAxes)  # Switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # Bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # Bottom-right diagonal

# Set labels and titles
ax2.set_xlabel('CPUs per Worker')
ax1.set_ylabel('Avg. Runtime per Unit')
ax2.set_ylabel('Avg. Runtime per Unit')
ax1.set_title('Example of a Broken Axis')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
cpus_per_worker = [8, 4, 2, 1]

# Runtimes for different configurations
runtime_200_pdfs = [87, 28, 28, 28]
runtime_23395_pdfs = [111, 54, 54, 54]
runtime_23395_zips = [459, 120, np.nan, np.nan]

# Constants
n_200 = 200
n_23395_pdfs = 23395
n_23395_zips = 23395

N_200 = 200
N_23395 = 23395

# Calculate Avg. runtime per (runtime / (N * cpus_per_worker))
avg_runtime_200_pdfs = [rt / (n_200) for rt, cpw in zip(runtime_200_pdfs, cpus_per_worker)]
avg_runtime_23395_pdfs = [rt / (n_23395_pdfs) for rt, cpw in zip(runtime_23395_pdfs, cpus_per_worker)]
avg_runtime_23395_zips = [rt / (n_23395_zips) if not np.isnan(rt) else np.nan for rt, cpw in zip(runtime_23395_zips, cpus_per_worker)]

# Create subplots with a break in the y-axis
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 8))

x = np.arange(len(cpus_per_worker))
width = 0.2

# Plot bars on both axes
ax1.bar(x - width, avg_runtime_200_pdfs, width, label='200 PDFs (not zipped)')
ax1.bar(x, avg_runtime_23395_pdfs, width, label='23,395 PDFs (256 PDFs/ZIP)')
ax1.bar(x + width, avg_runtime_23395_zips, width, label='23,395 ZIPs (512 PDFs/ZIP)')

ax2.bar(x - width, avg_runtime_200_pdfs, width)
ax2.bar(x, avg_runtime_23395_pdfs, width)
ax2.bar(x + width, avg_runtime_23395_zips, width)

# Set y-axis limits for the break
ax1.set_ylim(0.25, 0.45)  # Upper part
ax2.set_ylim(0, 0.03)  # Lower part

# Remove spines and ticks between the plots
ax1.spines['bottom'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax1.tick_params(labeltop=False)  # Don't put tick labels at the top of the upper plot
ax2.xaxis.tick_bottom()

# Add diagonal lines to indicate a break in the y-axis
d = .015  # How big to make the diagonal lines in axes coordinates
kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
ax1.plot((-d, +d), (-d, +d), **kwargs)        # Top-left diagonal
ax1.plot((1 - d, 1 + d), (-d, +d), **kwargs)  # Top-right diagonal

kwargs.update(transform=ax2.transAxes)  # Switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # Bottom-left diagonal
ax2.plot((1 - d, 1 + d), (1 - d, 1 + d), **kwargs)  # Bottom-right diagonal

# Labels and title
ax2.set_xlabel('CPUs per Worker')
ax1.set_ylabel('Avg. Runtime per Unit')
ax2.set_ylabel('Avg. Runtime per Unit')
ax1.set_title('Avg. Runtime per Unit vs. CPUs per Worker')

# Set x-ticks and labels
ax2.set_xticks(x)
ax2.set_xticklabels(cpus_per_worker)

# Add legend
ax1.legend()

# Add a caption-like text box
caption = "Note: The process failed for >256 PDFs (e.g., 512 and 1024 PDFs per ZIP)."
plt.figtext(0.5, -0.1, caption, wrap=True, horizontalalignment='center', fontsize=12)

# Display the plot
plt.show()
```
